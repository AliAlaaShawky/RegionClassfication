```python
import cv2
import numpy as np
import glob as gb
###sample of dataset "i used samle of dataset to low processing">>this sample in "test "folder
trainpath=r'C:/Users/xps/Desktop/task/test'
images=gb.glob(pathname=str(trainpath+'/*.png'))
idx = 0 
for img in images:
    idx=idx+1
    image = cv2.imread(img)
    gray=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY) 
    edged = cv2.Canny(image, 10, 250) 
    (cnts, _) = cv2.findContours(edged.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE) 
    for c in cnts: 
        x,y,w,h = cv2.boundingRect(c) 
        if w>50 and h>50: 
            
            new_img=image[y:y+h,x:x+w] 
            #cv2.imwrite('New/'+str(idx) + '.png', new_img)
            gray = cv2.cvtColor(new_img, cv2.COLOR_BGR2GRAY)
            ret, thresh = cv2.threshold(gray, 240, 255, cv2.THRESH_BINARY)
            cv2.bitwise_not(thresh, thresh)

            cnts = cv2.findContours(thresh, cv2.RETR_EXTERNAL,
                            cv2.CHAIN_APPROX_SIMPLE)
            cnts = cnts[0]
            c = max(cnts, key=cv2.contourArea)

            left = tuple(c[c[:, :, 0].argmin()][0])
            right = tuple(c[c[:, :, 0].argmax()][0])

            distance = np.sqrt( (right[0] - left[0])**2 + (right[1] - left[1])**2 )

            x,y,w,h = cv2.boundingRect(c)

            centx = np.sqrt( ((right[0] + left[0])**2)/4)
            centy = np.sqrt( ((right[1] + left[1])**2)/4 )
            print('>>>>>>>>x  :', centx)
            print('>>>>>>>>y  :', centy)
            font = cv2.FONT_HERSHEY_SIMPLEX
            cv2.circle(new_img, left, 5, (0, 0, 255), -1)
            cv2.circle(new_img, right, 5, (0, 0, 255), -1)
            cv2.circle(new_img, (int(centx), int(centy)), 5, (0, 0, 255), -1)
            cv2.line(new_img, left, right, (255,0,0), 2)
            cv2.drawContours(new_img, [c], -1, (0,255,0), 2)
            cv2.rectangle(new_img,(x,y),(x+w,y+h),(0,255,0),2)
            cv2.putText(new_img,'Distance: '+str(distance),(10,30), font, 1, (0,0,0),2, cv2.LINE_AA)
            print('>>>>>>distance :',distance)
            
            
```
